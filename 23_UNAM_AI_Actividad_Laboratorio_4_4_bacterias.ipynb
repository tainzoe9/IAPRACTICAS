# Bacterias

IMPLEMENTACIÓN DE NAIVE BAYES

```python
# Division de datos para entrenamientoy prueba
#Importamos la librería

from sklearn.model_selection import train_test_split

#Separo los datos de  "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#Entrenamiento

```python
#Entrenamiento con el 70% que son con los 280 registros
classifier.fit(X_train, y_train)
```

```python
# Prediccion del conjunto de prueba de acuerdo al 20% para los entrenamientos 120 registros
y_pred = classifier.predict(X_test)
```

# 1. Regresión Logística (Logistic Regression)
La regresión logística intentará encontrar un hiperplano (una línea, en este caso 2D) que separe las dos clases de bacterias.

¿Cómo funciona?: Calcula una combinación lineal de las entradas y le aplica la función sigmoide para obtener una probabilidad entre 0 y 1.

Limitación: Si las bacterias están distribuidas de forma circular o si una clase rodea a la otra (distribuciones no lineales), la Regresión Logística tendrá un desempeño pobre porque solo puede trazar fronteras rectas.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

#Entrenamiento del modelo de regresion logistica
log_reg=LogisticRegression()
log_reg.fit(X_train, y_train)

#Prediccion sobre el mismo set (como pide el ejercicio)
y_pred = log_reg.predict(X_train)
print("Reporte de Regresion Logistica")
print(classification_report(y_train, y_pred))
```

# 2. Máquinas de Vector de Soporte (SVM)
SVM es más robusto porque permite usar el Kernel Trick para encontrar fronteras de decisión complejas.

Kernel Lineal: Similar a la regresión logística.

Kernel RBF (Gaussiano): Crea fronteras curvas. Es muy probable que este sea el necesario para el dataset de bacterias si las concentraciones óptimas forman "nubes" o grupos cerrados.

```python
from sklearn.svm import SVC

#Entrenamiento con un Kernel RBF para capturar no linealidades
svm_model=SVC(kernel='rbf', gamma='auto')
svm_model.fit(X_train, y_train)

y_pred=svm_model.predict(X_train)
print("Reporte de SVM")
print(classification_report(y_train, y_pred))
```

#Visualización

```python
def plot_decision_boundary(model, X, y, title):
    h=0.2 # tamaño del paso en la malla
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
    plt.title(title)
    plt.show()
```

```python
plot_decision_boundary(log_reg, X_train, y_train, "Naive Bayes - Frontera de Decisión")
plot_decision_boundary(svm_model, X_train, y_train, "SVM - Frontera de Decisión")
```

#Conclusión

El clasificador Naive Bayes ingenuo mostró una precisión baja (aproximadamente 52.5%) en los datos proporcionados. Esto se debe a que Naive Bayes asume independencia entre las características, lo cual puede no ser cierto en este conjunto de datos.

Al visualizar la frontera de decisión, se puede observar que el clasificador Naive Bayes intenta crear una separación lineal simple, lo cual no es adecuado para datos que no son linealmente separables. La Regresión Logística, al ser también un clasificador lineal, obtuvo un rendimiento similar.

Por otro lado, el clasificador SVM con un kernel RBF (Gaussiano) logró un mejor rendimiento (aproximadamente 73% de precisión), demostrando su capacidad para aprender fronteras de decisión no lineales. Esto sugiere que las bacterias (B1 y B2) no están separadas linealmente por las concentraciones de los químicos X1 y X2, sino que sus regiones de supervivencia son más complejas o 'circulares' en el espacio de características.

En conclusión, para este tipo de datos, un clasificador más flexible como SVM con kernel RBF es más apropiado que Naive Bayes o Regresión Logística, ya que puede capturar mejor la relación no lineal entre las características y las clases.
