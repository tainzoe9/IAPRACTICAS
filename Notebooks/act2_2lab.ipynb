```python
# Crear una lista con al menos cinco elementos
lista_elementos = ["Elemento 1", "Elemento 2", "Elemento 3", "Elemento 4", "Elemento 5"]

# Imprimir la lista original
print("Lista original:")
print(lista_elementos)

# Agregar un nuevo elemento en el índice 2
nuevo_elemento = "Nuevo Elemento"
lista_elementos.insert(2, nuevo_elemento)

# Imprimir la lista después de agregar un nuevo elemento en el índice 2
print("\nLista después de agregar un nuevo elemento en el índice 2:")
print(lista_elementos)

# Agregar un elemento al final de la lista
elemento_final = "Elemento Final"
lista_elementos.append(elemento_final)

# Imprimir la lista después de agregar un elemento al final
print("\nLista después de agregar un elemento al final:")
print(lista_elementos)
```

```python
# Escribe una cadena de caracteres que contenga tres palabras
cadena = "Hola, ¿cómo estás?"

# Cuenta los caracteres de la cadena generada
long = len(cadena)

# Imprime el número de caracteres
print("El número de caracteres en la cadena es:", long)

# Guarda la cadena resultante toda en mayúsculas en una nueva variable
cadena_mayusculas = cadena.upper()

# Imprime la cadena en mayúsculas
print("La cadena en mayúsculas es:", cadena_mayusculas)
```

```python
# Escribe un diccionario que contenga al menos cuatro elementos
diccionario = {"a": 10, "b": 20, "c": 30, "d": 40}

# Imprime el diccionario original
print("Diccionario inicial:", diccionario)

# Agrega un nuevo elemento al diccionario
diccionario["nuevo"] = 50

# Imprime el diccionario después de agregar un nuevo elemento
print("Diccionario tras agregar 'nuevo':", diccionario)

# Elimina el elemento del índice "c"
if "c" in diccionario:
    del diccionario["c"]

# Imprime el diccionario después de eliminar el elemento 'c'
print("Diccionario tras eliminar 'c':", diccionario)

# Cuenta los elementos que componen el diccionario después de completadas las operaciones anteriores
num_elementos = len(diccionario)
print("El número total de elementos es:", num_elementos)
```

```python
# Solicitar la edad del usuario
edad = int(input("Ingresa tu edad: "))

# Calcular el costo de la entrada según la edad
if edad < 3:
    precio = 0
elif 3<= edad <= 12:
    precio = 10
else:
    precio = 15

# Mostrar el costo de la entrada al usuario
print("El costo de la entrada es:", precio)
```

```python
# Crear un diccionario con la información de la canción
cancion = {
    "artista": "Twenty One Pilots",
    "album": "Blurryface",
    "cancion": "Stressed Out",
    "duracion": "3:22",
    "genero": "Alternative Hip Hop"
}

# Mostrar las keys del diccionario
print("Keys del diccionario:")
print(cancion.keys())

# Mostrar los values del diccionario
print("\nValues del diccionario:")
print(cancion.values())

# Mostrar las tuplas (key, value) del diccionario
print("\nTuplas (key, value) del diccionario:")
print(cancion.items())
```
