```python
import csv
import json

# Función para generar un diccionario a partir de una lista de nombre, apellido y edad
def generar_diccionario_personas(lista_personas):
    diccionario_personas = []
    for persona in lista_personas:
        nombre, apellido, edad = persona
        diccionario_personas.append({"nombre": nombre, "apellido": apellido, "edad": edad})
    return diccionario_personas

# Lista de personas ficticias
lista_personas1 = [("Juan", "Pérez", 30), ("María", "Gómez", 25), ("Luis", "Martínez", 35)]
lista_personas2 = [("Ana", "López", 28), ("Pedro", "Rodríguez", 40), ("Sofía", "Fernández", 22)]
lista_personas3 = [("Carlos", "Sánchez", 50), ("Elena", "Díaz", 33), ("Ricardo", "Ruiz", 45)]

# Generar diccionarios
diccionario_personas1 = generar_diccionario_personas(lista_personas1)
diccionario_personas2 = generar_diccionario_personas(lista_personas2)
diccionario_personas3 = generar_diccionario_personas(lista_personas3)

# Almacenar los resultados en una lista de diccionarios
resultados = diccionario_personas1 + diccionario_personas2 + diccionario_personas3

# Escribir en un archivo CSV
csv_file_path = 'personas.csv'
keys = ['nombre', 'apellido', 'edad']

with open(csv_file_path, 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=keys)
    writer.writeheader()
    writer.writerows(resultados)

print(f"Data written to {csv_file_path}")

# Escribir en un archivo JSON
json_file_path = 'personas.json'

with open(json_file_path, 'w') as jsonfile:
    json.dump(resultados, jsonfile, indent=4)

print(f"Data written to {json_file_path}")

# Leer el contenido del archivo CSV
print(f"\nContent of {csv_file_path}:")
with open(csv_file_path, 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        print(row)

# Leer el contenido del archivo JSON
print(f"\nContent of {json_file_path}:")
with open(json_file_path, 'r') as jsonfile:
    json_content = json.load(jsonfile)
    print(json.dumps(json_content, indent=4))

# Definir la clase Persona
class Persona:
    def __init__(self, nombre, apellido, edad):
        self.nombre = nombre
        self.apellido = apellido
        self.edad = edad

# Crear una instancia de la clase Persona
mi_persona = Persona("Zoe", "Sotomayor", 22)
print("\nInformación de la instancia de Persona:")
print("Nombre:", mi_persona.nombre)
print("Apellido:", mi_persona.apellido)
print("Edad:", mi_persona.edad)
```
