Este código utiliza el algoritmo de poda alfa-beta en el juego del tic-tac-toe. Puedes probarlo tal como está.



https://www.youtube.com/watch?v=I0y-TGehf-4

```python
import math

def imprimir_tablero(tablero):
    for fila in tablero:
        print(" ".join(fila))
    print()

def es_tablero_lleno(tablero):
    for fila in tablero:
        if " " in fila:
            return False
    return True

def evaluar_estado(tablero):
    # Evaluación simple para el tic-tac-toe
    ganador = None
    # Filas
    for fila in tablero:
        if fila.count("X") == 3:
            ganador = "X"
        elif fila.count("O") == 3:
            ganador = "O"
    # Columnas
    for col in range(3):
        if tablero[0][col] == tablero[1][col] == tablero[2][col] and tablero[0][col] != " ":
            ganador = tablero[0][col]
    # Diagonales
    if tablero[0][0] == tablero[1][1] == tablero[2][2] and tablero[0][0] != " ":
        ganador = tablero[0][0]
    if tablero[0][2] == tablero[1][1] == tablero[2][0] and tablero[0][2] != " ":
        ganador = tablero[0][2]

    if ganador == "X":
        return 1
    elif ganador == "O":
        return -1
    elif es_tablero_lleno(tablero):
        return 0
    else:
        return None

def alphabeta(tablero, profundidad, alfa, beta, es_maximizando):
    estado_actual = evaluar_estado(tablero)

    if estado_actual is not None:
        return estado_actual

    if es_maximizando:
        mejor_valor = -math.inf
        for i in range(3):
            for j in range(3):
                if tablero[i][j] == " ":
                    tablero[i][j] = "X"
                    valor = alphabeta(tablero, profundidad + 1, alfa, beta, False)
                    tablero[i][j] = " "
                    mejor_valor = max(mejor_valor, valor)
                    alfa = max(alfa, mejor_valor)
                    if beta <= alfa:
                        break
        return mejor_valor
    else:
        mejor_valor = math.inf
        for i in range(3):
            for j in range(3):
                if tablero[i][j] == " ":
                    tablero[i][j] = "O"
                    valor = alphabeta(tablero, profundidad + 1, alfa, beta, True)
                    tablero[i][j] = " "
                    mejor_valor = min(mejor_valor, valor)
                    beta = min(beta, mejor_valor)
                    if beta <= alfa:
                        break
        return mejor_valor

def mejor_movimiento(tablero):
    mejor_valor = -math.inf
    mejor_mov = None
    alfa = -math.inf
    beta = math.inf

    for i in range(3):
        for j in range(3):
            if tablero[i][j] == " ":
                tablero[i][j] = "X"
                valor = alphabeta(tablero, 0, alfa, beta, False)
                tablero[i][j] = " "

                if valor > mejor_valor:
                    mejor_valor = valor
                    mejor_mov = (i, j)
                alfa = max(alfa, mejor_valor)

    return mejor_mov

# Ejemplo de uso con tic-tac-toe
tablero_inicial = [
    [" ", " ", " "],
    [" ", " ", " "],
    [" ", " ", " "]
]

turno = "X"

while evaluar_estado(tablero_inicial) is None and not es_tablero_lleno(tablero_inicial):
    imprimir_tablero(tablero_inicial)

    if turno == "X":
        fila, col = map(int, input("Turno de X. Ingrese fila y columna (separadas por espacio): ").split())
        if tablero_inicial[fila][col] == " ":
            tablero_inicial[fila][col] = "X"
        else:
            print("Casilla ocupada. Intenta de nuevo.")
            continue
    else:
        print("Turno de O (IA)")
        movimiento = mejor_movimiento(tablero_inicial)
        fila, col = movimiento
        tablero_inicial[fila][col] = "O"

    turno = "O" if turno == "X" else "X"

imprimir_tablero(tablero_inicial)
resultado = evaluar_estado(tablero_inicial)

if resultado == 1:
    print("¡X gana!")
elif resultado == -1:
    print("¡O gana!")
else:
    print("¡Empate!")
```

```python
import numpy as np
import random
import math
import time

# --- CONSTANTES ---
FILAS = 6
COLUMNAS = 7
JUGADOR = 1
IA = 2
VACIO = 0
VENTANA_LONGITUD = 4

# --- HERRAMIENTA DE MEDICIÓN ---
def medir_tiempo(func):
    def wrapper(*args, **kwargs):
        inicio = time.time()
        res = func(*args, **kwargs)
        fin = time.time()
        print(f"[Tiempo] {func.__name__}: {(fin-inicio)*1000:.2f} ms")
        return res
    return wrapper

# --- LÓGICA DEL JUEGO CONECTA 4 ---
def crear_tablero():
    return np.zeros((FILAS, COLUMNAS), dtype=int)

def es_valido(tablero, col):
    return tablero[FILAS-1][col] == 0

def obtener_fila_siguiente(tablero, col):
    for r in range(FILAS):
        if tablero[r][col] == 0:
            return r

def soltar_ficha(tablero, fila, col, pieza):
    tablero[fila][col] = pieza

def es_ganador(tablero, pieza):
    # Revisar horizontal, vertical y diagonales
    # (Implementación simplificada para brevedad)
    for c in range(COLUMNAS-3):
        for r in range(FILAS):
            if all(tablero[r][c+i] == pieza for i in range(4)): return True
    for c in range(COLUMNAS):
        for r in range(FILAS-3):
            if all(tablero[r+i][c] == pieza for i in range(4)): return True
    for c in range(COLUMNAS-3):
        for r in range(FILAS-3):
            if all(tablero[r+i][c+i] == pieza for i in range(4)): return True
            if all(tablero[r+3-i][c+i] == pieza for i in range(4)): return True
    return False

# --- HEURÍSTICA DE EVALUACIÓN ---
def evaluar_ventana(ventana, pieza):
    score = 0
    oponente = JUGADOR if pieza == IA else IA

    if ventana.count(pieza) == 4:
        score += 100
    elif ventana.count(pieza) == 3 and ventana.count(VACIO) == 1:
        score += 5
    elif ventana.count(pieza) == 2 and ventana.count(VACIO) == 2:
        score += 2

    if ventana.count(oponente) == 3 and ventana.count(VACIO) == 1:
        score -= 4 # Bloquear al oponente es prioritario

    return score

def evaluar_posicion(tablero, pieza):
    score = 0
    # Centro
    array_centro = [int(i) for i in list(tablero[:, COLUMNAS//2])]
    score += array_centro.count(pieza) * 3

    # Horizontal, Vertical, Diagonal
    # (Se recorre el tablero evaluando ventanas de 4 espacios)
    # ... [Se omite código repetitivo, la lógica usa evaluar_ventana] ...
    return score # Retorna un valor numérico estimado

# --- ALGORITMO MINIMAX CON PODA ALFA-BETA ---
def minimax(tablero, profundidad, alpha, beta, maximizando):
    es_terminal = es_ganador(tablero, JUGADOR) or es_ganador(tablero, IA) or len(obtener_movimientos_validos(tablero)) == 0

    if profundidad == 0 or es_terminal:
        if es_terminal:
            if es_ganador(tablero, IA): return (None, 1000000000)
            elif es_ganador(tablero, JUGADOR): return (None, -1000000000)
            else: return (None, 0) # Empate
        else:
            return (None, evaluar_posicion(tablero, IA)) # Heurística

    movimientos = obtener_movimientos_validos(tablero)
    if maximizando:
        valor = -math.inf
        columna = random.choice(movimientos)
        for col in movimientos:
            row = obtener_fila_siguiente(tablero, col)
            b_copy = tablero.copy()
            soltar_ficha(b_copy, row, col, IA)
            nuevo_score = minimax(b_copy, profundidad-1, alpha, beta, False)[1]
            if nuevo_score > valor:
                valor = nuevo_score
                columna = col
            alpha = max(alpha, valor)
            if alpha >= beta: break # PODA
        return columna, valor
    else: # Minimizando
        valor = math.inf
        columna = random.choice(movimientos)
        for col in movimientos:
            row = obtener_fila_siguiente(tablero, col)
            b_copy = tablero.copy()
            soltar_ficha(b_copy, row, col, JUGADOR)
            nuevo_score = minimax(b_copy, profundidad-1, alpha, beta, True)[1]
            if nuevo_score < valor:
                valor = nuevo_score
                columna = col
            beta = min(beta, valor)
            if alpha >= beta: break # PODA
        return columna, valor

def obtener_movimientos_validos(tablero):
    return [c for c in range(COLUMNAS) if es_valido(tablero, c)]

@medir_tiempo
def mejor_jugada_ia(tablero):
    col, minimax_score = minimax(tablero, 4, -math.inf, math.inf, True) # Profundidad 4
    return col

tablero = crear_tablero()
mejor_jugada_ia(tablero)
```
